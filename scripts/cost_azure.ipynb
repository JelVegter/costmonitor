```python
import json
import os
import pandas as pd
from typing import Union
pd.options.display.max_columns = None
```

## Setup

```python
data_app = [
    "caerus",
    "hermes",
    "styx",
]

environments = [
    "dev",
    "tst",
    "acc",
    "prd"
]

PRICE_COL = "costInPricingCurrency"

raw_data_directory = "data/raw/"
processed_data_directory = "data/processed/"
```

## Load data

```python

def load_files_to_df(directory: str) -> pd.DataFrame:
    """Function to load all files in a directory into a dataframe"""
    files = os.listdir(directory)
    files = [file.replace("/","_") for file in files if file[-4:] == ".csv"]
    df = pd.DataFrame()
    for file in files:
        df = pd.concat([
            df,
            pd.read_csv(directory+file, index_col=[0])
        ])
    return df
df = load_files_to_df(raw_data_directory)
df.shape
```

## Prep

```python
df["resourceGroupName"] = df["resourceGroupName"].str.lower()
```

```python
def get_keyword_match(value: str, keywords: list[str]) -> str:
        if isinstance(value, float):
            return value
            
        for keyword in keywords:
            if value.__contains__(keyword):
                return keyword

df["DataApp"] = df["resourceGroupName"].apply(lambda x: get_keyword_match(x, data_app))
df["Environment"] = df["resourceGroupName"].apply(lambda x: get_keyword_match(x, environments))
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
def get_resource_name(value: str) -> str:
    return value.split("/")[-1]

df["ResourceName"] = df["ResourceId"].map(get_resource_name)
```

```python
def get_databricks_cluster_name(value: Union[float,str]) -> Union[None,str]:

    if isinstance(value, str):
        tags = json.loads(value)
        if "ClusterName" in tags.keys():
            return tags["ClusterName"]
    return None


df["DatabricksCluster"] = df["tags"].map(get_databricks_cluster_name)
df["DatabricksCluster"]
```

## Filters

```python
anti_df = df.loc[~df["DataApp"].isin(data_app)]
print("resource groups left out: ", set(anti_df["resourceGroupName"].to_list()))
df = df.loc[df["DataApp"].isin(data_app)]
df.shape
```

## Calculations

```python
costs = df
```

```python
def estimate_DBU_costs(df: pd.DataFrame, multiplier: float) -> pd.DataFrame:
    dbu_estimation = df.loc[
        (df["meterCategory"] == "Virtual Machines")
        & (df["resourceGroupName"].str.contains("databricks"))
        ]

    dbu_estimation["meterCategory"] = "Azure Databricks"
    dbu_estimation["meterSubCategory"] = "Azure Databricks"
    dbu_estimation[PRICE_COL] = dbu_estimation[PRICE_COL] * multiplier

    return dbu_estimation

dbu = estimate_DBU_costs(df, 1.48)
costs = pd.concat([costs, dbu])
```

```python
costs.to_csv(processed_data_directory+"costs.csv")
```
